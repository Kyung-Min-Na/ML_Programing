S_All.txt와 Z_All.txt를 각각 이용, 512개씩 800개의 행렬 묶음을 생성.
1. 각 레벨(800개씩 나뉜거)의 절대값의평균
2. 각 레벨의 모든 계수들을 제곱하고 평균
3. 각 레벨 안에 있는 모든 계수들의 중앙값

```python
import numpy as np
```

```python
# 파일 경로 지정 / Google Colab
file_path_s = '/content/drive/MyDrive/Colab Notebooks/S_All.txt'
file_path_z = '/content/drive/MyDrive/Colab Notebooks/Z_All.txt'
```

```python
# data loader
def read_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
    data = [float(line.strip()) for line in lines]
    return data

# S_All.txt와 Z_All.txt 파일에서 data read
s_data = read_file(file_path_s)
z_data = read_file(file_path_z)
```

```python
# data split for 512 each
def divide_data_into_levels(data):
    levels = []
    for i in range(0, len(data), 512):
        level = data[i:i+512]
        levels.append(level)
    return levels

# 각 레벨의 작업 수행
def process_levels(levels):
    results = []
    for level in levels:
        # 1. 절대값의 평균 구하기
        abs_mean = np.mean(np.abs(level))

        # 2. 제곱 후 평균 구하기
        squared_mean = np.mean(np.square(level))

        # 3. 중앙값 구하기
        median = np.median(level)

        results.append((abs_mean, squared_mean, median))
    return results

# S_All.txt와 Z_All.txt의 data split for 512 each
s_levels = divide_data_into_levels(s_data)
z_levels = divide_data_into_levels(z_data)

# 각 레벨에서 작업 수행
s_results = process_levels(s_levels)
z_results = process_levels(z_levels)

# print output
for i, (s_result, z_result) in enumerate(zip(s_results, z_results)):
    print(f"Level {i + 1}:")
    print(f"S_All.txt - 절대값 평균: {s_result[0]}, 제곱 후 평균: {s_result[1]}, 중앙값: {s_result[2]}")
    print(f"Z_All.txt - 절대값 평균: {z_result[0]}, 제곱 후 평균: {z_result[1]}, 중앙값: {z_result[2]}")
    print()
```
